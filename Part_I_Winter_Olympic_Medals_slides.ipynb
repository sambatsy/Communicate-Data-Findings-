## (Gender Distribution)

> The following visualization illustrates the gender distribution; as we concluded, more medals are awarded to men. Additionally, an interesting fact we have discovered is that there are medals awarded to both genders in mixed. 

````python
def pie_chart(column):
    #quick function that visualizes a pie chart
    sorted_values = w_df[column].value_counts()
    pie_chart_colors=['#7BC8F6', '#FFFF14', '#C79FEF']
    number = '%.1f%%'
    w_df.groupby(column).size().plot(kind='pie',autopct=number, colors=pie_chart_colors)
    plt.title("The distrubition of Gender in Winter Olympics for a Century")
    plt.axis("equal")
    #figure(figsize=(20,20))
    fig = plt.gcf()
    fig.set_size_inches(20.5, 12.5)
    plt.rcParams.update({'font.size': 24})
pie_chart('Gender')
````

#### no legends are needed for univariate visualization

## (Top 5 Youngest Ages for Winter Olympic Medalists)

> The following visualization shows the frequency of the top 5 youngest ages for Winter Olympic medalists, leading to the conclusion that the youngest medalists are in their mid-teenage years (1417).

````python
#sources: seaborn documentation + pandas
#edge in displot source: https://datavizpyr.com/add-outine-to-histogram-in-seaborn/
sb.set(rc={'figure.figsize':(10.2, 8.2)})
sb.set(font_scale=1.3)
plt.xlabel("Youngest Winter Olympians")
plt.ylabel("Age Frequency")
plt.title('Top 5 Youngest Ages for Winter Olympic Medalists')
sb.distplot(w_df['Age of Athlete'].nsmallest(n=40),kde=False,bins=10,hist_kws=dict(edgecolor="black", linewidth=1.5))
````

### no legends are needed for univariate visualization

## (Top 5 Countries and Their Most Common medals)

>When it comes to European countries, gold medals are most common, except for Austria (bronze medals are most common there).

````python
def bi_countplot(column,n):
    sb.set(font_scale=1.7)
    sb.set(rc={'figure.figsize':(12.7,8.27)})
    sb.countplot(x=column,data=w_df,hue='Medal',order=w_df[column].value_counts().head(n).index)
    plt.title('Top 5 Countries and Their Most Common medals')
bi_countplot('Country',5)
````
